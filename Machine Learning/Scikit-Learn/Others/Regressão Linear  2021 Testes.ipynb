```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
dados = {'Nome':['Camila','Julia','Thais','Marcelo','Felipe','João','Claudia','Dionoro'],
        'Peso':[50,55,45,70,75,120,80,100],
        'Altura':[160,165,158,170,180,190,175,185],
        'Sexo':['F','F','F','M','M','M','F','M']}
```

```python
df = pd.DataFrame(dados)
```

```python
df
```

# Peso

```python
sns.displot(df['Peso'],kind='kde');
```

# Altura

```python
sns.displot(df['Altura'],kind='kde');
```

## Scatter

```python
sns.relplot(x='Peso',y='Altura',data=df,hue='Sexo',palette=['red','blue']);
```

## Modelo linear

```python
sns.regplot(x='Peso',y='Altura',data=df,ci=90, truncate=False);
```

```python
df
```

```python
X = df[['Peso']]
Y = df[['Altura']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2 )
```

## Criando Modelo

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```
