# Basic Commands

* docker ps: list the active containers 
* docker ps -a: inactive containers too
* docker container run ubuntu: creating a container
* docker container stop "id": stop the specified container
* docker container logs -f "id": show the logs of the container
* docker container exec "id" _command_: execute a command in a container 
* docker container rm id: remove a container
* docker container prune: remove all inactives containers
* docker image rmi "id": remove the especified image of a container
* docker container start "id": start a container
* docker container run -it "id": conecte the terminal from the host to the terminal in container


# DockerFile
* FROM: install the image (ex: FROM node)
* ENV: Set tem enviroment variables
* RUN: Run linux commands in the container (ex: RUN echo "hello world")
* COPY: Copy something from your host to the container
* CMD: Excute a command (ex: CMD ["python", "app.py"]), it's a entry point
