# Request - GET - POST - PUT
<p>O request é uma biblioteca do python capaz de gerar solicitações http, assim como um usuário ou seja extremamente útil para manipulação de APIs.</p>

* GET
* POST
* PUT

<p>Basicamente com ele podemos fazer 3 tipos de solicitações, "GET" = apenas leitura, "POST" e "PUT" = para modificar os dados.</p>
<p>Utilizando o método GET</p>

```python
import requests
res = requests.get('https://scotch.io') 
```

<p>Geralmente quando a solicitação for efetuada corretamente isso se resultará em um output = 200, para o print(res).</p>

```python
res
```

<p>Podemos então utilizar comandos como o headers para receber o output de um cabeçalho de uma determinada página, com os determinados tipos de dados.</p>

```python
res.headers['Date']
```

<p>E o text para analisarmos o conteúdo do código HTML.</p>

```python
res.text
```

# Extraindo Dados

<p>Podemos além de requisitar links, passar parâmetros para eles, como no seguinte.</p> 

```python
params = {'i':'tt0137523'}
res = requests.get('http://www.omdbapi.com/?apikey=b92f9408', params = params)
```

<p>O params, fornece os parâmetros de entrada da url.</p>

```python
res.url
```

<p>Podemos ainda transformar a str, em um objeto json, para facilitar a manipulação.</p>

```python
var = res.json()
var
```

<p><b>Dica:
Quando vemos um URL com “e” comercial (&), pontos de interrogação (? ) e sinais de igual (=), é certo que o URL é feito para pedidos de GET. Esses símbolos especificam os parâmetros que acompanham o URL.</b></p>
