```python
import pandas as pd
```

# Utilização Do Merge - .merge()

## inner Join

<p> Quando utilizamos o merge sem definir como será a união, ele sempre optará pela união <b>Inner</b>, essa união faz com que os dados se unam apenas com dados que ambos DataFrames tenham. </p>



## left join ou right join
<p> Quando utilizarmos nossos joins com a opção <b>how = 'left' ou how = 'right'</b> estaremos optando por definir um DataFrame como prioridade sendo o da esquerda ou o da direita, nesse caso será acrescentado dados do DataFrame secundário ao DataFrame prioritário. </p>



## parametro left_on e right_on 

<p> Será necessário definir esse parametro quando os DataFrames da direita e esquerda tiverem como coluna merge nomes diferentes. </p>



## outer join
<p> Quando definimos os outer joins, nós pegaremos todos os dados de ambos DataFrames, sem definir um prioritário </p>



## pd.merge_ordered()
<p> O merge ordered tem como default os seguinte parametro how = 'outer' </p>
<p> Ao realizar a união o merge_ordered() fara a união dos dataframes da mesma maneira que o merge tradicional, porém ele pode receber um parametro chamado <b>fill_method = 'ffill'</b> (foward fill) que acrescentará o valor anterior a todos os dados posteriores que eram NaN </p>

# Concatenação De Dados - .concat()

## Criando nossos dados de exemplo

```python
conc = {
    'ID':[1,2,3,4,5],
    'MÊS':['JAN','FEV','MAR','ABR','MAI']
}
conc2 = {
    'ID':[6,7,8,9,10],
    'MÊS':['JUN','JUL','AGO','SET','OUT']
}
conc1 = pd.DataFrame(conc)
conc2 = pd.DataFrame(conc2)
```

```python
conc1
```

```python
conc2
```

## .concat()
<p> Fazendo junção de dois dataframes mesmo que com colunas diferentes </p>

```python
pd.concat([conc1,conc2], ignore_index = True)
```

### parametro keys, criando chaves para o DF

```python
pd.concat([conc1,conc2], ignore_index = False, keys = ['df1','df2'])
```

### parametro verify_integrity - removendo dados duplicados

```python
pd.concat([conc1,conc2], ignore_index = False, keys = ['df1','df2'], verify_integrity = True)
```
