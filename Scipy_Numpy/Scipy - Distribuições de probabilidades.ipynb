## Skew (coeficiente de assimetria de pearson)
<p> O skew mede a assimetria dos dados </p>

* Se a direita o skew será positivo
* Se a esquerda o skew será esquerdo
* Se for 0 simetrico

```python
from scipy.stats import skew
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Arrays perdidos
a = np.array([1,2,2,3,3,3,4,4,5])
b = np.array([1,2,2,2,3,4,5,6,7])
c = np.array([1,2,3,4,5,6,6,6,7])

print(skew(a),'-', skew(b), '-', skew(c))
```

```python
lista_skews = [a,b,c]

for i in lista_skews:
    sns.displot(i, kde=True)
    plt.title(f'Skew: {i}')
    plt.show()
```

## kurtosis( )
<p> Ao sabermos o grau de <b>K</b> poderemos definir a distribuição dos dados. </p>

```python
from scipy.stats import kurtosis
```

```python
distrib_nrm = np.random.normal(1,3,100000000)
plt.hist(distrib_nrm,bins=50)
plt.show()
```

```python
kurtosis(distrib_nrm) # OS DADOS ESTÃO CONCENTRADOS AO REDOR DA MEDIA
```

```python
distrib_logist = np.random.logistic(10, 1, 100)
plt.hist(distrib_logist,bins=50)
plt.show()
```

```python
kurtosis(distrib_logist)
```

```python
distrib_gamma = np.random.gamma(2.,2.,100) # OS DADOS ESTÃO DISPERSOS PA KRL
plt.hist(distrib_gamma,bins=50)
plt.show()
```

```python
kurtosis(distrib_gamma)
```

# Distribuições

## Distribuição normal
Suponha que você queira determinar a probabilidade de que um cliente tenha um saldo bancário maior que $100.000,00, dado que a média de saldo bancário dos clientes é de $75.000,00 e o desvio padrão é de $10.000,00.



```python
from scipy.stats import norm

# Probabilidade de que um cliente tenha um saldo bancário maior que $100.000,00
prob = 1 - norm.cdf(100000, loc=75000, scale=10000)
print(prob)
```

* Cerca de 68,26% de todos os dados estão dentro de um desvio padrão ( <σ ) da média (μ)
* Cerca de 95,44% de todos os dados estão dentro de dois desvios padrão ( 2σ ) da média (μ)
* Cerca de 99,72% de todos os dados estão dentro de três desvios padrão ( <3σ ) da média (μ)
* E os 0,28% restantes de todos os dados estão fora de três desvios padrão ( >3σ ) da média (μ)
* Os dados fora de três desvios padrão é considerada como outliers .





## Poisson
Suponha que você queira determinar a probabilidade de que exatamente 3 componentes falhem em um circuito eletrônico em um período de 24 horas, sabendo que a taxa média de falhas é de 2 componentes por dia.

```python
from scipy.stats import poisson

# Probabilidade de que exatamente 3 componentes falhem em um circuito eletrônico em um período de 24 horas
prob = poisson.pmf(3, 2*24)
print(prob)
```

## Binomial
Suponha que você queira determinar o número mínimo de usuários que podem ser atendidos por uma equipe de suporte de 10 pessoas para garantir que menos de 5% dos usuários fiquem na fila de espera por mais de 5 minutos. Suponha que a probabilidade de um usuário precisar de mais de 5 minutos de atendimento seja de 20%.

```python
from scipy.stats import binom

# Número mínimo de usuários que podem ser atendidos simultaneamente para garantir que menos de 5% dos usuários fiquem na fila de espera por mais de 5 minutos
n = 1
while binom.cdf(n, 100, 0.2) > 0.05:
    n += 1
print(n)
```

## Distribuição exponencial

Suponha que você queira determinar o tempo médio de espera dos clientes para agendar uma entrega, sabendo que a taxa média de chegada dos clientes é de 10 por hora.

```python
from scipy.stats import expon

# Tempo médio de espera dos clientes para agendar uma entrega
espera_media = expon.mean(scale=1/10)
print(espera_media)
```
