# Parametros Importantes graficos Relacionais e Categoricos
<p> Com <b>relplot()</b> podemos criar gráficos de relação entre duas variáveis, como gráfico de linhas e disperssão </p>



* Parametros
* Alpha: Regulariza o grau de desfoque;
* Hue: Aplica uma coluna categorica para separar nossos dados;



* Palette: Modifica a paleta de cores, com base em um dicionário ou uma paleta padrão;
* Hue_Order: Cria uma ordenação dos dados com base em uma lista;
* Style: Aplica uma coluna categorica para estilizar nossos dados de maneira diferente;



* Col: Aplica uma coluna categorica para separar nosso gráfico em dois por coluna.
* Row: Funciona como Col porém, separa os gráficos por linha.



* Col_Wrap: Define quantos gráficos haverão por linha;
* Col_Order: Cria uma ordenação dos gráficos com base em uma lista;

<p> O hue também pode ser utilizado em outros tipos de gráficos </p>



* Kind: Seleciona o tipo de gráfico, para relplot temos ('line','scatter').



* Markers: Adiciona marcadores ao gráfico, caso style esteja ativado, os marcadores se apresentam de maneira diferente;
* ci: Representa o intervalo de confiança, pode ser utilizado None para sua eliminação ou 'sd' para usar o desvio padrão.



* Size: Modifica o tamanho do gráfico com base em uma coluna;



* Countplot: é um gráfico que faz a contagem da quantidade de repetições de variaveis em um conjunto.



* Kind: Temos no catplot os seguintes tipos de kind: ["strip”, “swarm”, “box”, “violin”, “boxen”, “point”, “bar”, or “count”];
* Sym: Se colocado em branco, representa a eliminação dos outliers no gráfico;



* Whis: representa, como os gráficos boxplot serão exibidos, exemplo [0,100] exibirá o ponto minimo e o máximo.



* Join: Representa a união com uma linha entre os dados;
* Estimator: representa o estimador que será avaliado;

## Disperssão De Dados Categoricos



* Stripplot: Cria uma distribuição do volume dos dados de acordo com suas quantidades;
* Jitter: o paramêtro jitter ajuda a evitar a sobreposição dos dados;
* Swarmplot: exibe todos os pontos sem sobreposição;

# Histogramas



* Distplot: trás consigo a possibilidade de plotagem de 3 tipos de gráficos, KDE (gráfico de densindade), Histograma e ECDF; 

<p><b> Distplot agora é chamado de displot </b></p>

# Regressão



* Regplot: Faz uma plotagem de uma regressão;
* Residplot: Faz uma plotagem do gráfico de resíduos;
* Order: Modifica o grua do polinomio da regressão;
* X_bins: Divide os dados numericos de maneira categorica;
* Xy_jitter: Acrescenta ruido, sendo util quando é feito tentativa da plotagem com dados categoricos;
* Xy_estimator: Acrescenta um estimador, como a media, mediana etc, para representar os pontos do determinado eixo;



* Lmplot: Funciona de maneira parecida ao Regplot, porém utiliza de maior desempenho por ser mais costumizado com: Col;Markers;Hue.

## Heat Map



* Heatmap: Proporciona um gráfico de calor, mostrando a frequencia dos dados conforme a cor;
* Annot: Exibe os valores de cada dado;
* Cbar: Acrescenta ou remove a barra do lado direito;
* Linewidths: Cria e aumenta as linhas entre os dados;

<p><b>Podemos utilizar o heatmap junto ao gráfico de correlação.</b></p>

# Primeira Visualização



* Pairplot: Exibe o grafico de cada um de nossos dados com suas determinadas distribuições e relações, podemos utilizar, kind=reg entre outros parametros;
* Jointplot: Exibe a relação dos nossos dados com suas determinadas distribuições;
* Hex: Apresenta nossa relação de maneira Hexagonal;

# Estilização



* Set_style: Modifica o estilo do gráfico;
* Set_palette: Modifica a paletra de cores do gráfico;
* Set_context: Formata o estilo de exibição do gráfico;



<p> Necessário <b> g = catplot() ou relplot() </b></p>

* g.fig.suptitle: Coloca um titulo no gráfico;
* g.set(xlabel =, ylabel=): Acrescenta as labels ao gráfico;
* plt.xticks(rotation=): Pode ser usado para efetuar rotação dos eixos;
