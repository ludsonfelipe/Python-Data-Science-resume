# Parametros Importantes graficos Relacionais e Categoricos
<p> Com <b>relplot()</b> podemos criar grficos de relao entre duas variveis, como grfico de linhas e dispersso </p>



* Parametros
* Alpha: Regulariza o grau de desfoque;
* Hue: Aplica uma coluna categorica para separar nossos dados;



* Palette: Modifica a paleta de cores, com base em um dicionrio ou uma paleta padro;
* Hue_Order: Cria uma ordenao dos dados com base em uma lista;
* Style: Aplica uma coluna categorica para estilizar nossos dados de maneira diferente;



* Col: Aplica uma coluna categorica para separar nosso grfico em dois por coluna.
* Row: Funciona como Col porm, separa os grficos por linha.



* Col_Wrap: Define quantos grficos havero por linha;
* Col_Order: Cria uma ordenao dos grficos com base em uma lista;

<p> O hue tambm pode ser utilizado em outros tipos de grficos </p>



* Kind: Seleciona o tipo de grfico, para relplot temos ('line','scatter').



* Markers: Adiciona marcadores ao grfico, caso style esteja ativado, os marcadores se apresentam de maneira diferente;
* ci: Representa o intervalo de confiana, pode ser utilizado None para sua eliminao ou 'sd' para usar o desvio padro.



* Size: Modifica o tamanho do grfico com base em uma coluna;



* Countplot:  um grfico que faz a contagem da quantidade de repeties de variaveis em um conjunto.



* Kind: Temos no catplot os seguintes tipos de kind: ["strip, swarm, box, violin, boxen, point, bar, or count];
* Sym: Se colocado em branco, representa a eliminao dos outliers no grfico;



* Whis: representa, como os grficos boxplot sero exibidos, exemplo [0,100] exibir o ponto minimo e o mximo.



* Join: Representa a unio com uma linha entre os dados;
* Estimator: representa o estimador que ser avaliado;

## Dispersso De Dados Categoricos



* Stripplot: Cria uma distribuio do volume dos dados de acordo com suas quantidades;
* Jitter: o paramtro jitter ajuda a evitar a sobreposio dos dados;
* Swarmplot: exibe todos os pontos sem sobreposio;

# Histogramas



* Distplot: trs consigo a possibilidade de plotagem de 3 tipos de grficos, KDE (grfico de densindade), Histograma e ECDF; 

<p><b> Distplot agora  chamado de displot </b></p>

# Regresso



* Regplot: Faz uma plotagem de uma regresso;
* Residplot: Faz uma plotagem do grfico de resduos;
* Order: Modifica o grua do polinomio da regresso;
* X_bins: Divide os dados numericos de maneira categorica;
* Xy_jitter: Acrescenta ruido, sendo util quando  feito tentativa da plotagem com dados categoricos;
* Xy_estimator: Acrescenta um estimador, como a media, mediana etc, para representar os pontos do determinado eixo;



* Lmplot: Funciona de maneira parecida ao Regplot, porm utiliza de maior desempenho por ser mais costumizado com: Col;Markers;Hue.

## Heat Map



* Heatmap: Proporciona um grfico de calor, mostrando a frequencia dos dados conforme a cor;
* Annot: Exibe os valores de cada dado;
* Cbar: Acrescenta ou remove a barra do lado direito;
* Linewidths: Cria e aumenta as linhas entre os dados;

<p><b>Podemos utilizar o heatmap junto ao grfico de correlao.</b></p>

# Primeira Visualizao



* Pairplot: Exibe o grafico de cada um de nossos dados com suas determinadas distribuies e relaes, podemos utilizar, kind=reg entre outros parametros;
* Jointplot: Exibe a relao dos nossos dados com suas determinadas distribuies;
* Hex: Apresenta nossa relao de maneira Hexagonal;

# Estilizao



* Set_style: Modifica o estilo do grfico;
* Set_palette: Modifica a paletra de cores do grfico;
* Set_context: Formata o estilo de exibio do grfico;



<p> Necessrio <b> g = catplot() ou relplot() </b></p>

* g.fig.suptitle: Coloca um titulo no grfico;
* g.set(xlabel =, ylabel=): Acrescenta as labels ao grfico;
* plt.xticks(rotation=): Pode ser usado para efetuar rotao dos eixos;
